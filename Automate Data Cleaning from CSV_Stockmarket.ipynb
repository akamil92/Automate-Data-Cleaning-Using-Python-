```python
#Started data cleaning

## Add a new column to the DataFrame
df['Shareholder TYPES(new sort)'] = ''

# Iterate over the rows in the DataFrame and update the new column based on the Investor_Type column
for index, row in df.iterrows():
    investor_type = row['Investor_Type']
    if investor_type in ["RETAIL BODY CORPORATES - PRIVATE LIMITED COMPANY", "CORPORATE BODY - BANK/FINANCE COMPANY", "RETAIL BODY CORPORATES - PRIVATE LIMITED COMPANY", "CORPORATE BODY - CO-OPERATIVES"]:
        df.at[index, 'Shareholder TYPES(new sort)'] = 'CORPORATE'
    elif investor_type == "RETAIL BODY CORPORATES - LIMITED COMPANY":
        df.at[index, 'Shareholder TYPES(new sort)'] = 'CORPORATE'
    elif investor_type in ["INSTITUTIONAL BODY CORPORATES - LIMITED COMPANY", "CORPORATE BODY - INVESTMENT TRUST/FOUNDATION", "OTHERS"]:
        df.at[index, 'Shareholder TYPES(new sort)'] = 'INSTITUTIONAL'
    else:
        df.at[index, 'Shareholder TYPES(new sort)'] = 'INDIVIDUAL'

# Print the updated DataFrame to confirm the new column was added and updated

df
```

```python
# Add new columns to the DataFrame
df['Pledge(Broker)'] = ''
df['Pledge(Shareholder)'] = ''

# Iterate over the rows in the DataFrame and update the new columns based on the Qualifier column
for index, row in df.iterrows():
    qualifier = row['Qualifier']
    if isinstance(qualifier, float) and np.isnan(qualifier):
        continue
    pledge_parts = str(qualifier).split('FOR')
    if len(pledge_parts) == 2:
        df.at[index, 'Pledge(Broker)'] = pledge_parts[0].strip()
        df.at[index, 'Pledge(Shareholder)'] = pledge_parts[1].strip()
    else:
        df.at[index, 'Pledge(Broker)'] = pledge_parts[0].strip()
```

```python
df
```

```python
#copy 'Pledge(Shareholder)' in to 'Combine Name (CDS + NOMINEE)
df['Combine Name (CDS + NOMINEE)'] = df['Pledge(Shareholder)']
```

```python
df
```

```python
# Copy data from Qualifier column to 'Combine Name (CDS + NOMINEE) column for empty cells only
df['Combine Name (CDS + NOMINEE)'] = df['Combine Name (CDS + NOMINEE)'].replace('', pd.NA).fillna(df['Qualifier'])
```

```python
# Save the updated DataFrame as a CSV file
df.to_csv(r'C:\Users\dell\Desktop\New dashboard , combine companies\dataset\Agmo\AGMO NO Pass new\automated data clean\untransfer automated clean data.csv', index=False)

```

```python
# Copy data from Holder_Name column to 'Combine Name (CDS + NOMINEE) column for empty cells only
df['Combine Name (CDS + NOMINEE)'] = df['Combine Name (CDS + NOMINEE)'].replace('', pd.NA).fillna(df['Holder_Name'])
```

```python
df
```

```python
# Update the Pledge(Shareholder) column
df['Pledge(Shareholder)'] = df['Pledge(Shareholder)'].str.replace(r'\(.*', '', regex=True)
```

```python
#do gender colunn
for index, row in df.iterrows():
    investor_type = row['Investor_Type']
    investor_id = row['Investor_ID']
    if investor_type == 'INDIVIDUAL':
        if investor_id[-1].isnumeric():
            last_digit = int(investor_id[-1])
            if last_digit % 2 == 0:
                df.at[index, 'Gender'] = 'FEMALE'
            else:
                df.at[index, 'Gender'] = 'MALE'
        else:
            continue
```

```python
# do age column 
df['Age'] = ''

# Iterate over the rows in the DataFrame and update the new column based on the Investor_ID column
for index, row in df.iterrows():
    investor_type = row['Investor_Type']
    investor_id = row['Investor_ID']
    if investor_type == 'INDIVIDUAL':
        if investor_id[:2].isdigit():
            first_two_digits = int(investor_id[:2])
            if 20 <= first_two_digits <= 99:
                df.at[index, 'Age'] = 2023 - (1900 + first_two_digits)
            elif 0 <= first_two_digits <= 20:
                df.at[index, 'Age'] = 2023 - (2000 + first_two_digits)
            else:
                continue
        else:
            continue
```

```python
df['Combine Name (CDS + NOMINEE)'] = df['Combine Name (CDS + NOMINEE)'].str.replace(r'\(.*\)', '', regex=True)
```

```python
# Save the updated DataFrame as a CSV file
df.to_csv(r'C:\Users\dell\Desktop\New dashboard , combine companies\dataset\Agmo\AGMO NO Pass new\automated data clean\before do IVT.csv', index=False)

```

```python

# Replace IVT with data from Holder_Name and Pledge(Broker) in Combine Name (CDS + NOMINEE) column
for index, row in df.iterrows():
    if 'IVT' in row['Combine Name (CDS + NOMINEE)']:
        holder_name = row['Holder_Name']
        pledge_broker = row['Pledge(Broker)']
        new_name = holder_name + ' (' + pledge_broker + ')'
        df.at[index, 'Combine Name (CDS + NOMINEE)'] = new_name
```
